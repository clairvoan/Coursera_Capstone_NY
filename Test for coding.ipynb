```python
import pandas as pd
url="https://raw.githubusercontent.com/clairvoan/Coursera_Capstone_NY/master/Calgaryschool.csv"
c=pd.read_csv(url)
c.head()
```

```python
url2="https://raw.githubusercontent.com/clairvoan/Coursera_Capstone_NY/master/Edmontonschool.csv"
e=pd.read_csv(url2)
e.head()
```

```python
c.shape
```

```python
e.shape
```

```python
dfcal = c.drop('Rating in the most recent five years', axis=1)
dfcal.head()
```

```python
dfedm = e.drop('Rating in the most recent five years', axis=1)
dfedm.head()
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X = dfcal.values[:,1:]
X = np.nan_to_num(X)
Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet
```

```python
clusterNum = 5
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
dfcal["Clus_km"] = labels
dfcal.head(5)
```

```python
dfcal.groupby('Clus_km').mean()
```

```python
Y = dfedm.values[:,1:]
Y = np.nan_to_num(Y)
Clus_dataSet1 = StandardScaler().fit_transform(Y)
Clus_dataSet1
```

```python
clusterNum1 = 5
k_means1 = KMeans(init = "k-means++", n_clusters = clusterNum1, n_init = 12)
k_means1.fit(Y)
labels1 = k_means1.labels_
print(labels1)
```

```python
dfedm["Clus_km1"] = labels1
dfedm.head(5)
```

```python
dfedm.groupby('Clus_km1').mean()
```

```python
colordictcal = {0: 'yellow', 1: 'green', 2: 'red', 3: 'lightgreen', 4: 'orange'}
```

```python
latitude_cal = 51.044270
longitude_cal = -114.062019
cal_map = folium.Map(location=[latitude_cal, longitude_cal], zoom_start=12)

cal_map
```

```python
for lat, lon, schoolzone in zip(dfcal['latitude'], dfcal['Longitude'], dfcal['Clus_km']):
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        color='b',
        key_on = schoolzone,
        threshold_scale=[0,1,2,3,4],
        fill_color=colordictcal[schoolzone],
        fill=True,
        fill_opacity=0.7
        ).add_to(cal_map)
cal_map
```
